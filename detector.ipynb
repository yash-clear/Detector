```python
import numpy as np
import  cv2

```

```python

from IPython.display import Image
detect=cv2.imread('/content/scene3.jpg')
height=detect.shape[0]
width=detect.shape[1]
resized=cv2.resize(detect,(300,300))
img_blob=cv2.dnn.blobFromImage(resized,0.007843,(300,300),127.5)
Image('/content/scene3.jpg')
```

```python
classNames = { 0: 'background',
    1: 'aeroplane', 2: 'bicycle', 3: 'bird', 4: 'boat',
    5: 'bottle', 6: 'bus', 7: 'car', 8: 'cat', 9: 'chair',
    10: 'cow', 11: 'diningtable', 12: 'dog', 13: 'horse',
    14: 'motorbike', 15: 'person', 16: 'pottedplant',
    17: 'sheep', 18: 'sofa', 19: 'train', 20: 'tvmonitor' }
labels=list(classNames.values())
```

```python
mobilenetssd=cv2.dnn.readNetFromCaffe('/content/mobilenetssd.prototext','/content/mobilenetssd.caffemodel')
mobilenetssd.setInput(img_blob)
obj_detection=mobilenetssd.forward()
```

```python
from google.colab.patches import cv2_imshow
total_detection=obj_detection.shape[2]
for i in np.arange(0, total_detection):
  predicted=obj_detection[0,0,i,2]
  if predicted>0.2:
    predicted_index=int(obj_detection[0,0,i,1])
    predicted_label=labels[predicted_index]

    box=obj_detection[0,0,i,3:7]*np.array([width,height,width,height])
    (startX, startY, endX, endY) = box.astype("int")

    predicted_label = "{}: {:.2f}%".format(labels[predicted_index], predicted*100)
    print("predicted object {}:{}".format(i+1,predicted_label))
    cv2.rectangle(detect,(startX, startY),( endX, endY),(255,0,200),1)
    cv2.putText(detect, predicted_label, (startX, startY-5),cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0,0,255), 1)
cv2_imshow(detect)
```

```python
import numpy as np
import  cv2
```

```python

from IPython.display import Image
detect=cv2.imread('/content/scene4.jpg')
height=detect.shape[0]
width=detect.shape[1]
#resized=cv2.resize(detect,(300,300))
img_blob=cv2.dnn.blobFromImage(detect,crop=False)
Image('/content/scene4.jpg')
```

```python
labels=["person","bicycle","car","motorbike","aeroplane","bus","train","truck","boat","traffic light","fire hydrant","street sign","stop sign","parking meter","bench","bird","cat","dog","horse","sheep","cow","elephant","bear","zebra","giraffe","hat","backpack","umbrella","shoe","eye glasses","handbag","tie","suitcase","frisbee","skis","snowboard","sports ball","kite","baseball bat","baseball glove","skateboard","surfboard","tennis racket","bottle","plate","wine glass","cup","fork","knife","spoon","bowl","banana","apple","sandwich","orange","broccoli","carrot","hot dog","pizza","donut","cake","chair","sofa","pottedplant","bed","mirror","diningtable","window","desk","toilet","door","tv","laptop","mouse","remote","keyboard","cell phone","microwave","oven","toaster","sink","refrigerator","blender","book","clock","vase","scissors","teddy bear","hair drier","toothbrush"]
```

```python

color=["0,255,0","255,0,0","0,0,255","255,255,0","0,255,255","255,0,255"]

color=[np.array(every_color.split(",")).astype("int") for  every_color in color]
color=np.array(color)

maskrcnn=cv2.dnn.readNetFromTensorflow('/content/maskrcnn_buffermodel.pb','/content/maskrcnn_bufferconfig.txt')
maskrcnn.setInput(img_blob)
(obj_detection_box,obj_detection_mask)= maskrcnn.forward(["detection_out_final","detection_masks"])
```

```python
from google.colab.patches import cv2_imshow
import random

total_detection=obj_detection_box.shape[2]
for i in np.arange(0, total_detection):
  predicted=obj_detection_box[0,0,i,2]
  if predicted>0.5:
    predicted_index=int(obj_detection_box[0,0,i,1])
    predicted_label=labels[predicted_index]

    box=obj_detection_box[0,0,i,3:7]*np.array([width,height,width,height])
    (startX, startY, endX, endY) = box.astype("int")
    
    box_width=endX-startX
    box_height=endY-startY
    mask=obj_detection_mask[i,predicted_index]
    mask=cv2.resize(mask,(box_width,box_height))
    mask=(mask>0.3)
    region=detect[startY:endY,startX:endX]
    region=region[mask]
    mask_color=random.choice(color)

    cover=((0.3*mask_color)+(0.5*region)).astype('uint8')
    detect[startY:endY,startX:endX][mask]=cover
    mask_color=[int(x) for x in mask_color]


    predicted_label = "{}: {:.2f}%".format(labels[predicted_index], predicted*100)
    print("predicted object {}:{}".format(i+1,predicted_label))
    cv2.rectangle(detect,(startX, startY),( endX, endY),(255,120,200),1)
    cv2.putText(detect, predicted_label, (startX, startY-5),cv2.FONT_HERSHEY_SIMPLEX, 0.4, (0,0,255), 1)
cv2_imshow(detect)

```
